### 1.  Analyze store sales:

```python
# Calculate total sales per store
total_sales = store_data.groupby('Store_Id')['Store_Sales'].sum()
total_sales
```

```python
# Calculate total sales per store
total_sales = store_data.groupby('Items_Available')['Store_Sales'].sum()

# Visualize total sales per store
total_sales.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Items_Available')
plt.ylabel('Total Sales (US $)')
plt.title('Total Sales per Store')
plt.show()
```

```python
# Calculate total sales per store
total_sales = store_data.groupby('Items_Available')['Store_Sales'].sum()

# Get top 10 stores based on total sales
top_10_stores = total_sales.nlargest(10)

# Visualize top 10 stores (with red bars)
top_10_stores.plot(kind='bar', figsize=(10, 6), color='red')
plt.xlabel('Items_Available')
plt.ylabel('Total Sales (US $)')
plt.title('Top 10 Stores by Total Sales')
plt.show()
```

```python
Observation
This analysis provides insights into the performance of different supermarket store branches based on their total sales.
The visualization helps identify the top-performing stores and can be used to make informed business decisions.
```

###  2.  Analyze customer count:

```python
# Calculate average daily customer count per store
avg_customer_count = store_data.groupby('Store_Id')['Daily_Customer_Count'].mean()
avg_customer_count
```

```python
# Calculate average daily customer count per store
avg_customer_count = store_data.groupby('Store_Id')['Daily_Customer_Count'].mean()

# Get top 10 stores based on average daily customer count
top_10_stores = avg_customer_count.nlargest(10)

# Visualize top 10 stores (with orange bars)
top_10_stores.plot(kind='bar', figsize=(10, 6), color='orange')
plt.xlabel('Store_Id')
plt.ylabel('Average Daily Customer Count')
plt.title('Top 10 Stores by Average Daily Customer Count')
plt.show()
```

```python
Observation
This analysis provides insights into the popularity and customer engagement of different supermarket store ssbranches based
on their average daily customer count. The visualization helps identify the stores with the highest customer footfall and
can be used to make strategic decisions related to staffing, promotions,
and customer service.
```

### 3. Analyze store area and items available:

```python
# Calculate total store area and items available per store
total_area = store_data.groupby('Store_Id')['Store_Area'].sum()
total_items = store_data.groupby('Store_Id')['Items_Available'].sum()

# Visualize total store area and items available per store
fig, ax = plt.subplots(2, 1, figsize=(10, 8))
total_area.plot(kind='bar', ax=ax[0])
ax[0].set_xlabel('Store_Id')
ax[0].set_ylabel('Total Store Area (yard square)')
ax[0].set_title('Total Store Area per Store')

total_items.plot(kind='bar', ax=ax[1])
ax[1].set_xlabel('Store_Id')
ax[1].set_ylabel('Total Items Available')
ax[1].set_title('Total Items Available per Store')

plt.tight_layout()
plt.show()
```
